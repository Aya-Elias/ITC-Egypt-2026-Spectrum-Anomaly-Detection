```python
# 03_results_analysis.ipynb
# تحليل نتائج نموذج تصنيف إشارات RF - ITC-EGYPT 2026

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import json
import os

print("="*60)
print("📊 تحليل نتائج النموذج - EfficientNet-B0")
print("="*60)

# ============================================
# 1. تحميل النموذج والتقرير
# ============================================
print("\n[1] تحميل النموذج والتقرير...")

MODEL_PATH = "/kaggle/working/best_model.keras"
REPORT_PATH = "/kaggle/working/model_report.json"

model = tf.keras.models.load_model(MODEL_PATH)
with open(REPORT_PATH, 'r') as f:
    report = json.load(f)

print("✅ تم التحميل")
```

```python
# ============================================
# 2. عرض التقرير
# ============================================
print("\n[2] تقرير النموذج:")
print(f"   الدقة الكلية: {report['accuracy']:.4f}")
print(f"   عدد عينات الاختبار: {report['test_samples']}")
print(f"   توزيع العينات: {report['class_distribution_test']}")

print("\n📋 تقرير التصنيف لكل فئة:")
for class_name, metrics in report['classification_report'].items():
    print(f"   {class_name}: Precision={metrics['precision']:.2f}, Recall={metrics['recall']:.2f}, F1={metrics['f1']:.2f}")

# ============================================
# 3. مصفوفة الارتباك (Confusion Matrix)
# ============================================
print("\n[3] مصفوفة الارتباك...")

cm = np.array(report['confusion_matrix'])
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Drone', 'Normal', 'Jamming'],
            yticklabels=['Drone', 'Normal', 'Jamming'])
plt.title('Confusion Matrix - EfficientNet-B0 (93.47%)')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150)
plt.show()
```

```python
# ============================================
# 4. تحليل الأخطاء
# ============================================
print("\n[4] تحليل الأخطاء:")

# الأخطاء من المصفوفة
drone_errors = cm[0][1] + cm[0][2]  # Drone اتحسب Normal أو Jamming
normal_errors = cm[1][0] + cm[1][2]  # Normal اتحسب Drone أو Jamming
jamming_errors = cm[2][0] + cm[2][1]  # Jamming اتحسب Drone أو Normal

print(f"   أخطاء Drone: {drone_errors} من {np.sum(cm[0])} ({drone_errors/np.sum(cm[0])*100:.1f}%)")
print(f"   أخطاء Normal: {normal_errors} من {np.sum(cm[1])} ({normal_errors/np.sum(cm[1])*100:.1f}%)")
print(f"   أخطاء Jamming: {jamming_errors} من {np.sum(cm[2])} ({jamming_errors/np.sum(cm[2])*100:.1f}%)")

# ============================================
# 5. الخلاصة
# ============================================
print("\n[5] الخلاصة:")
print("   ✅ Normal و Jamming: دقة 100% Recall")
print("   ✅ Drone: 80% Recall (أقل بسبب تعدد أنواع الإشارات)")
print(f"   ✅ الدقة الكلية: {report['accuracy']:.4f}")
print("   ✅ النموذج جاهز للمسابقة")
```

```python
# ============================================
# 6. حفظ التحليل
# ============================================
analysis_report = {
    "model": "EfficientNet-B0",
    "accuracy": report['accuracy'],
    "total_errors": int(drone_errors + normal_errors + jamming_errors),
    "error_breakdown": {
        "Drone": int(drone_errors),
        "Normal": int(normal_errors),
        "Jamming": int(jamming_errors)
    },
    "analysis_complete": True
}

with open("results_analysis.json", "w") as f:
    json.dump(analysis_report, f, indent=2)

print("\n✅ تم حفظ تحليل النتائج في results_analysis.json")
print("\n" + "="*60)
print("✅ اكتمل تحليل النتائج")
print("="*60)
```
