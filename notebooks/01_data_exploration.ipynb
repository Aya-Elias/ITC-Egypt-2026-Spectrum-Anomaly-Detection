```python
# 01_data_exploration.ipynb
# استكشاف البيانات الأولية لمشروع تصنيف إشارات RF (ITC-EGYPT 2026)

import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

print("="*60)
print("🔍 استكشاف البيانات - ITC-EGYPT 2026")
print("="*60)
```

```python
# ============================================
# 1. بيانات Drone (RFUAV)
# ============================================
print("\n[1] بيانات Drone (RFUAV)")

drone_path = "/kaggle/working/Drone_Processed"

if os.path.exists(drone_path):
    files = os.listdir(drone_path)
    print(f"   ✅ عدد الصور: {len(files)}")

    # فحص عينة
    sample = np.load(os.path.join(drone_path, files[0]))
    print(f"   ✅ Shape: {sample.shape}")
    print(f"   ✅ Dtype: {sample.dtype}")
    print(f"   ✅ Min: {sample.min():.2f}, Max: {sample.max():.2f}")

    # رسم عينة
    plt.figure(figsize=(6,6))
    plt.imshow(sample)
    plt.title("Sample Drone Spectrogram")
    plt.axis('off')
    plt.show()
else:
    print("   ❌ المسار غير موجود")
```

```python
# ============================================
# 2. بيانات Normal (RadioML)
# ============================================
print("\n[2] بيانات Normal (RadioML)")

normal_path = "/kaggle/working/Normal_Processed"

if os.path.exists(normal_path):
    files = os.listdir(normal_path)
    print(f"   ✅ عدد الصور: {len(files)}")

    sample = np.load(os.path.join(normal_path, files[0]))
    print(f"   ✅ Shape: {sample.shape}")
    print(f"   ✅ Dtype: {sample.dtype}")
    print(f"   ✅ Min: {sample.min():.2f}, Max: {sample.max():.2f}")
else:
    print("   ❌ المسار غير موجود")
```

```python
# ============================================
# 3. بيانات Jamming
# ============================================
print("\n[3] بيانات Jamming")

jamming_path = "/kaggle/working/Jamming_Processed"

if os.path.exists(jamming_path):
    files = os.listdir(jamming_path)
    print(f"   ✅ عدد الصور: {len(files)}")

    sample = np.load(os.path.join(jamming_path, files[0]))
    print(f"   ✅ Shape: {sample.shape}")
    print(f"   ✅ Dtype: {sample.dtype}")
    print(f"   ✅ Min: {sample.min():.2f}, Max: {sample.max():.2f}")
else:
    print("   ❌ المسار غير موجود")
```
